```python
# Ïù¥Î≤§Ìä∏ ÌÉÄÏûÖ Í∏∞Ï§Ä ÌïÑÌÑ∞ÎßÅ
views     = df_total[df_total['event_type'] == 'view']
purchases = df_total[df_total['event_type'] == 'purchase']

# Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Î∞©Î¨∏ Ïàò Î∞è Íµ¨Îß§ Ïàò
visit_counts    = views['category_lvl1'].value_counts().sort_index()
purchase_counts = purchases['category_lvl1'].value_counts().sort_index()

# Ïπ¥ÌÖåÍ≥†Î¶¨ Ïù¥Î¶Ñ ÌÜµÏùº
categories = sorted(set(visit_counts.index).union(set(purchase_counts.index)))

visit_counts = visit_counts.reindex(categories, fill_value=0)
purchase_counts = purchase_counts.reindex(categories, fill_value=0)

# ÎπÑÏú® Í≥ÑÏÇ∞ (%)
visit_rate    = 100 * visit_counts / visit_counts.sum()
purchase_rate = 100 * purchase_counts / purchase_counts.sum()

# ÏãúÍ∞ÅÌôî (subplot ÏñëÏòÜ)
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Î∞©Î¨∏Ïú® ÎßâÎåÄÍ∑∏ÎûòÌîÑ
axes[0].barh(categories, visit_rate, color='#a6c4e8')
axes[0].set_title('Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Î∞©Î¨∏Ïú®')
axes[0].set_xlabel('ÎπÑÏú® (%)')
axes[0].invert_yaxis()  # ÎÜíÏùÄ ÎπÑÏú® ÏúÑÎ°ú

# Íµ¨Îß§Ïú® ÎßâÎåÄÍ∑∏ÎûòÌîÑ
axes[1].barh(categories, purchase_rate, color='#4068e0')
axes[1].set_title('Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú®')
axes[1].set_xlabel('ÎπÑÏú® (%)')

# Ïä§ÌÉÄÏùº Í∞úÏÑ†
for ax in axes:
    ax.grid(axis='x', linestyle='--', alpha=0.5)
    ax.set_xlim(0, max(visit_rate.max(), purchase_rate.max()) * 1.2)

plt.tight_layout()
plt.show()
```

## üìà Ï£ºÎ†•ÏÉÅÌíà EDA

### üíµ 1Î≤àÎßå Íµ¨Îß§

### üîéÏπ¥ÌÖåÍ≥†Î¶¨ Î†àÎ≤®3Î≥Ñ Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù(Ï†ÑÌôòÏú®)

```python
# 0. event_time ÏÉùÏÑ±
top_one_df['event_time'] = pd.to_datetime(top_one_df['event_date']) + pd.to_timedelta(top_one_df['event_hour'], unit='h')

# 1. category_label ÏÉùÏÑ±
top_one_df['category_label'] = top_one_df['category_lvl1'] + ' > ' + top_one_df['category_lvl3']

# 2. Ïú†Ï†Ä + Ïπ¥ÌÖåÍ≥†Î¶¨ Îã®ÏúÑÎ°ú Ïù¥Î≤§Ìä∏ Ï†ïÎ†¨
top_one_df['event_datetime'] = top_one_df['event_time']
top_one_df_sorted = top_one_df.sort_values(by=['user_id', 'category_label', 'event_datetime'])

# 3. Ïú†Ï†Ä Îã®ÏúÑÎ°ú ÏàúÏÑú Í∏∞Î∞ò Ï†ÑÌôòÏú® Í≥ÑÏÇ∞ Ìï®Ïàò Ï†ïÏùò
def calculate_ordered_conversion_rate(df):
    user_events = df.groupby(['user_id', 'category_label'])['event_type'].apply(list).reset_index()

    results = []

    for _, row in user_events.iterrows():
        events = row['event_type']
        category = row['category_label']

        view_to_cart = 0
        cart_to_purchase = 0
        view_to_purchase = 0

        if 'view' in events and 'cart' in events:
            if events.index('view') < events.index('cart'):
                view_to_cart = 1
        if 'cart' in events and 'purchase' in events:
            if events.index('cart') < events.index('purchase'):
                cart_to_purchase = 1
        if 'view' in events and 'purchase' in events:
            if events.index('view') < events.index('purchase'):
                view_to_purchase = 1

        results.append({
            'category_label': category,
            'view_to_cart': view_to_cart,
            'cart_to_purchase': cart_to_purchase,
            'view_to_purchase': view_to_purchase
        })

    result_df = pd.DataFrame(results)

    # Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ï†ÑÌôòÏú® ÌèâÍ∑† (Ïú†Ï†Ä Îã®ÏúÑ)
    conversion_stats = result_df.groupby('category_label').mean().reset_index()
    return conversion_stats

# 4. Ï†ÑÌôòÏú® Í≥ÑÏÇ∞
conversion_stats = calculate_ordered_conversion_rate(top_one_df_sorted)

# 5. Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ïú†Ï†Ä Ïàò ÌïÑÌÑ∞ÎßÅ (30Î™Ö Ïù¥ÏÉÅ Î≥∏ Ïπ¥ÌÖåÍ≥†Î¶¨Îßå)
user_category_counts = top_one_df.groupby(['category_label', 'user_id']).size().reset_index(name='event_count')
user_counts_per_category = user_category_counts.groupby('category_label')['user_id'].nunique().reset_index(name='user_count')

# Ï†ÑÌôòÏú® + Ïú†Ï†Ä Ïàò Í≤∞Ìï©
conversion_stats = conversion_stats.merge(user_counts_per_category, on='category_label', how='left')

# ÌïÑÌÑ∞ÎßÅ
conversion_stats_filtered = conversion_stats[conversion_stats['user_count'] >= 30]

# Ï†ïÎ†¨
top_funnel_sorted = conversion_stats_filtered.sort_values(by='cart_to_purchase', ascending=False).head(10)
top_funnel_sorted

# 6. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(14, 6))
bar_width = 0.25
x = range(len(top_funnel_sorted))

plt.bar(x, top_funnel_sorted['view_to_cart'], width=bar_width, label='View ‚Üí Cart')
plt.bar([p + bar_width for p in x], top_funnel_sorted['cart_to_purchase'], width=bar_width, label='Cart ‚Üí Purchase')
plt.bar([p + bar_width*2 for p in x], top_funnel_sorted['view_to_purchase'], width=bar_width, label='View ‚Üí Purchase')

plt.xticks([p + bar_width for p in x], top_funnel_sorted['category_label'], rotation=45)
plt.ylabel('Ï†ÑÌôòÏú® (Ïú†Ï†Ä Í∏∞Ï§Ä)')
plt.title('Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù (Ïú†Ï†Ä + Ïù¥Î≤§Ìä∏ ÏàúÏÑú Í∏∞Î∞ò)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 0. event_time ÏÉùÏÑ±
top_one_df['event_time'] = pd.to_datetime(top_one_df['event_date']) + pd.to_timedelta(top_one_df['event_hour'], unit='h')

# 1. categoryÎ≥Ñ ÏßëÍ≥Ñ (view ‚Üí cart / view ‚Üí purchase)
views = top_one_df[top_one_df['event_type'] == 'view']
carts = top_one_df[top_one_df['event_type'] == 'cart']
purchases = top_one_df[top_one_df['event_type'] == 'purchase']

# category label ÏÉùÏÑ±
views['category_label'] = views['category_lvl1'] + ' > ' + views['category_lvl3']
carts['category_label'] = carts['category_lvl1'] + ' > ' + carts['category_lvl3']
purchases['category_label'] = purchases['category_lvl1'] + ' > ' + purchases['category_lvl3']

view_counts = views.groupby('category_label').size().reset_index(name='view_count')
cart_counts = carts.groupby('category_label').size().reset_index(name='cart_count')
purchase_counts = purchases.groupby('category_label').size().reset_index(name='purchase_count')

category_stats = view_counts.merge(cart_counts, on='category_label', how='outer')
category_stats = category_stats.merge(purchase_counts, on='category_label', how='outer')
category_stats.fillna(0, inplace=True)

category_stats['view_to_cart_rate'] = category_stats['cart_count'] / category_stats['view_count']
category_stats['view_to_purchase_rate'] = category_stats['purchase_count'] / category_stats['view_count']

# 2. cart ‚Üí purchase Ï†ÑÌôòÏú® (user_id + category Í∏∞Ï§Ä)
carts_time = carts.sort_values('event_time').groupby(['user_id', 'category_label'], as_index=False).first()
purchases_time = purchases.sort_values('event_time').groupby(['user_id', 'category_label'], as_index=False).first()

cart_purchase = pd.merge(carts_time, purchases_time, on=['user_id', 'category_label'], suffixes=('_cart', '_purchase'))
cart_purchase = cart_purchase[cart_purchase['event_time_purchase'] > cart_purchase['event_time_cart']]
converted_cart_purchase = cart_purchase[['user_id', 'category_label']].drop_duplicates()
converted_cart_purchase['to_purchase'] = True

carts_time['to_purchase'] = carts_time.merge(
    converted_cart_purchase,
    on=['user_id', 'category_label'],
    how='left'
)['to_purchase'].fillna(False).astype(bool)

cart_to_purchase_rate = carts_time.groupby('category_label')['to_purchase'].mean().reset_index()
cart_to_purchase_rate.columns = ['category_label', 'cart_to_purchase_rate']

# 3. ÌÜµÌï©
category_stats = pd.merge(category_stats, cart_to_purchase_rate, on='category_label', how='left')
category_stats.fillna(0, inplace=True)

# 4. ÌïÑÌÑ∞ÎßÅ Î∞è Ï†ïÎ†¨
category_stats_filtered = category_stats[category_stats['view_count'] >= 30]
top_funnel_sorted = category_stats_filtered.sort_values(by='view_to_purchase_rate', ascending=False).head(10)

# 5. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(14, 6))
bar_width = 0.25
x = range(len(top_funnel_sorted))

plt.bar(x, top_funnel_sorted['view_to_cart_rate'], width=bar_width, label='View ‚Üí Cart')
plt.bar([p + bar_width for p in x], top_funnel_sorted['cart_to_purchase_rate'], width=bar_width, label='Cart ‚Üí Purchase')
plt.bar([p + bar_width*2 for p in x], top_funnel_sorted['view_to_purchase_rate'], width=bar_width, label='View ‚Üí Purchase')

plt.xticks([p + bar_width for p in x], top_funnel_sorted['category_label'], rotation=45)
plt.ylabel('Ï†ÑÌôòÏú®')
plt.title('Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù (Ïπ¥ÌÖåÍ≥†Î¶¨ Í∏∞Ï§Ä ÏßëÍ≥Ñ)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 0. event_time ÏÉùÏÑ±
top_multi_df['event_time'] = pd.to_datetime(top_multi_df['event_date']) + pd.to_timedelta(top_multi_df['event_hour'], unit='h')

# 1. category_label ÏÉùÏÑ±
top_multi_df['category_label'] = top_multi_df['category_lvl1'] + ' > ' + top_multi_df['category_lvl3']

# 2. Ïú†Ï†Ä + Ïπ¥ÌÖåÍ≥†Î¶¨ Îã®ÏúÑÎ°ú Ïù¥Î≤§Ìä∏ Ï†ïÎ†¨
top_multi_df['event_datetime'] = top_multi_df['event_time']
top_multi_df_sorted = top_multi_df.sort_values(by=['user_id', 'category_label', 'event_datetime'])

# 3. Ïú†Ï†Ä Îã®ÏúÑÎ°ú ÏàúÏÑú Í∏∞Î∞ò Ï†ÑÌôòÏú® Í≥ÑÏÇ∞ Ìï®Ïàò Ï†ïÏùò
def calculate_ordered_conversion_rate(df):
    user_events = df.groupby(['user_id', 'category_label'])['event_type'].apply(list).reset_index()

    results = []

    for _, row in user_events.iterrows():
        events = row['event_type']
        category = row['category_label']

        view_to_cart = 0
        cart_to_purchase = 0
        view_to_purchase = 0

        if 'view' in events and 'cart' in events:
            if events.index('view') < events.index('cart'):
                view_to_cart = 1
        if 'cart' in events and 'purchase' in events:
            if events.index('cart') < events.index('purchase'):
                cart_to_purchase = 1
        if 'view' in events and 'purchase' in events:
            if events.index('view') < events.index('purchase'):
                view_to_purchase = 1

        results.append({
            'category_label': category,
            'view_to_cart': view_to_cart,
            'cart_to_purchase': cart_to_purchase,
            'view_to_purchase': view_to_purchase
        })

    result_df = pd.DataFrame(results)

    # Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ï†ÑÌôòÏú® ÌèâÍ∑† (Ïú†Ï†Ä Îã®ÏúÑ)
    conversion_stats = result_df.groupby('category_label').mean().reset_index()
    return conversion_stats

# 4. Ï†ÑÌôòÏú® Í≥ÑÏÇ∞
conversion_stats = calculate_ordered_conversion_rate(top_multi_df_sorted)

# 5. Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ïú†Ï†Ä Ïàò ÌïÑÌÑ∞ÎßÅ (30Î™Ö Ïù¥ÏÉÅ Î≥∏ Ïπ¥ÌÖåÍ≥†Î¶¨Îßå)
user_category_counts = top_multi_df.groupby(['category_label', 'user_id']).size().reset_index(name='event_count')
user_counts_per_category = user_category_counts.groupby('category_label')['user_id'].nunique().reset_index(name='user_count')

# Ï†ÑÌôòÏú® + Ïú†Ï†Ä Ïàò Í≤∞Ìï©
conversion_stats = conversion_stats.merge(user_counts_per_category, on='category_label', how='left')

# ÌïÑÌÑ∞ÎßÅ
conversion_stats_filtered = conversion_stats[conversion_stats['user_count'] >= 30]

# Ï†ïÎ†¨
top_funnel_sorted = conversion_stats_filtered.sort_values(by='cart_to_purchase', ascending=False).head(10)
top_funnel_sorted

# 6. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(14, 6))
bar_width = 0.25
x = range(len(top_funnel_sorted))

plt.bar(x, top_funnel_sorted['view_to_cart'], width=bar_width, label='View ‚Üí Cart')
plt.bar([p + bar_width for p in x], top_funnel_sorted['cart_to_purchase'], width=bar_width, label='Cart ‚Üí Purchase')
plt.bar([p + bar_width*2 for p in x], top_funnel_sorted['view_to_purchase'], width=bar_width, label='View ‚Üí Purchase')

plt.xticks([p + bar_width for p in x], top_funnel_sorted['category_label'], rotation=45)
plt.ylabel('Ï†ÑÌôòÏú® (Ïú†Ï†Ä Í∏∞Ï§Ä)')
plt.title('Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù (Ïú†Ï†Ä + Ïù¥Î≤§Ìä∏ ÏàúÏÑú Í∏∞Î∞ò)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 1. Í∞ÄÍ≤© Íµ¨Í∞Ñ Ï†ïÏùò
bins = [0, 100, 200, 300, 400, 500, 1000, float('inf')]
labels = ['0~100', '100~200', '200~300', '300~400', '400~500', '500~1000', '1000+']
top_multi_df['price_bin'] = pd.cut(top_multi_df['price'], bins=bins, labels=labels, right=False)

# 2. Ïú†Ï†Ä Îã®ÏúÑ Ï§ëÎ≥µ Ï†úÍ±∞
view_df = top_multi_df[top_multi_df['event_type'] == 'view'][['user_id', 'product_id', 'price_bin']].drop_duplicates()
purchase_df = top_multi_df[top_multi_df['event_type'] == 'purchase'][['user_id', 'product_id', 'price_bin']].drop_duplicates()

# 3. Í∞ÄÍ≤© Íµ¨Í∞ÑÎ≥Ñ Ïú†Ï†Ä Ïàò ÏßëÍ≥Ñ (FutureWarning Î∞©ÏßÄ ÏúÑÌï¥ observed=True Ï∂îÍ∞Ä)
view_stats = view_df.groupby('price_bin', observed=True)['user_id'].nunique().reset_index()
view_stats.columns = ['price_bin', 'view_user_count']

purchase_stats = purchase_df.groupby('price_bin', observed=True)['user_id'].nunique().reset_index()
purchase_stats.columns = ['price_bin', 'purchase_user_count']

# 4. Î≥ëÌï© Î∞è Ï†ÑÌôòÏú® Í≥ÑÏÇ∞
price_group_stats = pd.merge(view_stats, purchase_stats, on='price_bin', how='outer')

# ‚û§ Ïà´Ïûê Ïª¨ÎüºÎßå fillna(0)
price_group_stats['view_user_count'] = price_group_stats['view_user_count'].fillna(0)
price_group_stats['purchase_user_count'] = price_group_stats['purchase_user_count'].fillna(0)

price_group_stats['conversion_rate'] = price_group_stats['purchase_user_count'] / price_group_stats['view_user_count']
price_group_stats = price_group_stats.sort_values(by='price_bin')

# 5. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
plt.bar(price_group_stats['price_bin'].astype(str), price_group_stats['conversion_rate'])
plt.title('Í∞ÄÍ≤© Íµ¨Í∞ÑÎ≥Ñ Ï†ÑÌôòÏú® (Ïú†Ï†Ä Í∏∞Ï§Ä)')
plt.xlabel('Í∞ÄÍ≤© Íµ¨Í∞Ñ (ÎßåÏõê)')
plt.ylabel('Ï†ÑÌôòÏú®')
plt.tight_layout()
plt.show()
```

```python
# Ï≤´ Íµ¨Îß§ Ïù¥Ï†Ñ cart -> purchase ÏãúÍ∞Ñ Í≥ÑÏÇ∞
# 1. event_datetime Ïª¨Îüº ÏÉùÏÑ±
top_multi_df['event_datetime'] = top_multi_df['event_date'] + pd.to_timedelta(top_multi_df['event_hour'], unit='h')
top_multi_df['event_datetime'] = pd.to_datetime(top_multi_df['event_datetime'])

# 2. Í∞Å Ïú†Ï†ÄÏùò Ï≤´ purchase ÏãúÍ∞Ñ Íµ¨ÌïòÍ∏∞
first_purchase_time = top_multi_df[top_multi_df['event_type'] == 'purchase'] \
                        .groupby('user_id')['event_datetime'].min() \
                        .rename('first_purchase_time')

# 3. Ï≤´ Íµ¨Îß§ Ïù¥Ï†ÑÏùò ÌñâÎèôÎßå ÌïÑÌÑ∞ÎßÅ
top_multi_df = top_multi_df.join(first_purchase_time, on='user_id')
top_multi_df_first_purchase = top_multi_df[top_multi_df['event_datetime'] <= top_multi_df['first_purchase_time']].copy()

# 4. Ìï¥Îãπ Íµ¨Í∞Ñ ÎÇ¥ÏóêÏÑú cart/purchase ÏãúÏ†ê Ï∂îÏ∂ú
df_cart = top_multi_df_first_purchase[top_multi_df_first_purchase['event_type'] == 'cart']
df_purchase = top_multi_df_first_purchase[top_multi_df_first_purchase['event_type'] == 'purchase']

cart_times = df_cart.groupby('user_id')['event_datetime'].min()
purchase_times = df_purchase.groupby('user_id')['event_datetime'].min()

# 5. Í≥µÌÜµ user_id Í∏∞Ï§Ä Î≥ëÌï© Î∞è ÏãúÍ∞Ñ Ï∞® Í≥ÑÏÇ∞ (Îã®ÏúÑ: ÏãúÍ∞Ñ)
merged = pd.merge(cart_times, purchase_times, on='user_id', suffixes=('_cart', '_purchase'))
cart_hold_time_first_purchase = (merged['event_datetime_purchase'] - merged['event_datetime_cart']).dt.total_seconds() / 3600
```

```python
# Íµ¨Îß§ Ïù¥Î≤§Ìä∏Îßå ÌïÑÌÑ∞ÎßÅ
multi_purchase = top_multi_df[top_multi_df['event_type'] == 'purchase'].copy()

# ÎÇ†Ïßú Ï†ïÎ†¨
multi_purchase['event_date'] = pd.to_datetime(multi_purchase['event_date'])
multi_purchase = multi_purchase.sort_values(by=['user_id', 'event_date'])

# Ïú†Ï†ÄÎ≥Ñ Íµ¨Îß§ ÏàúÏÑú Î∂ÄÏó¨
multi_purchase['purchase_order'] = multi_purchase.groupby('user_id').cumcount() + 1

# Ï≤´ Î≤àÏß∏, Îëê Î≤àÏß∏ Íµ¨Îß§Îßå Ï∂îÏ∂ú
first_second = multi_purchase[multi_purchase['purchase_order'] <= 2]

# Ï≤´/Îëê Î≤àÏß∏ Íµ¨Îß§Î•º ÌïòÎÇòÏùò ÌñâÏóê ÌîºÎ≤ó
pivoted = first_second.pivot(index='user_id', columns='purchase_order', values='category_lvl1')
pivoted.columns = ['first_cat1', 'second_cat1']
pivoted = pivoted.dropna()

# Î≥ÄÍ≤Ω Ïó¨Î∂Ä ÌåêÎã®
pivoted['category_changed'] = pivoted['first_cat1'] != pivoted['second_cat1']

# ÌÜµÍ≥Ñ ÏöîÏïΩ
change_rate = pivoted['category_changed'].mean()
change_counts = pivoted['category_changed'].value_counts()

print("2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ÏûêÏùò Ï≤´‚ÜíÎëêÎ≤àÏß∏ Íµ¨Îß§ÏóêÏÑú Ïπ¥ÌÖåÍ≥†Î¶¨ Î≥ÄÍ≤Ω ÎπÑÏú®:", round(change_rate, 3))
print("\nÏÉÅÏÑ∏ Î∂ÑÌè¨:")
print(change_counts)

# Î≥ÄÍ≤ΩÎêú Í≤ΩÏö∞Îßå Ï∂îÏ∂ú
changed_cases = pivoted[pivoted['category_changed']]

# Î≥ÄÌôî Í≤ΩÎ°ú ÏßëÍ≥Ñ
change_paths = changed_cases.groupby(['first_cat1', 'second_cat1']).size().reset_index(name='count')
change_paths = change_paths.sort_values('count', ascending=False)

# ÏÉÅÏúÑ Î™á Í∞úÎßå ÏãúÍ∞ÅÌôî
top_paths = change_paths.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_paths, y='count', x='first_cat1', hue='second_cat1')
plt.title('2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ÏûêÏùò Ïπ¥ÌÖåÍ≥†Î¶¨ Î≥ÄÍ≤Ω Í≤ΩÎ°ú (Top 10)')
plt.xlabel('Ï≤´ Î≤àÏß∏ Íµ¨Îß§ Ïπ¥ÌÖåÍ≥†Î¶¨')
plt.ylabel('Î≥ÄÍ≤ΩÎêú Ïàò')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### üîé 'electronics' 'appliances' 'construction'Î≥Ñ ÌèâÍ∑† cart Î≥¥Ïú† ÏãúÍ∞Ñ ÎπÑÍµê

```python
# Î∂ÑÏÑù ÎåÄÏÉÅ Ïπ¥ÌÖåÍ≥†Î¶¨
target_categories = ['electronics', 'appliances', 'construction']

one_time_target = top_one_df[top_one_df['category_lvl1'].isin(target_categories)].copy()
multi_time_target = top_multi_df[top_multi_df['category_lvl1'].isin(target_categories)].copy()

# Cart Î≥¥Ïú† ÏãúÍ∞Ñ Í≥ÑÏÇ∞ Ìï®Ïàò
def calculate_cart_hold_hours(df, category_name):
    # Ìï¥Îãπ Ïπ¥ÌÖåÍ≥†Î¶¨ ÌïÑÌÑ∞ÎßÅ
    category_df = df[df['category_lvl1'] == category_name].copy()

    # event_datetime ÏÉùÏÑ±
    category_df['event_datetime'] = category_df['event_date'] + pd.to_timedelta(category_df['event_hour'], unit='h')
    category_df['event_datetime'] = pd.to_datetime(category_df['event_datetime'])

    # Ïû•Î∞îÍµ¨Îãà, Íµ¨Îß§ ÏãúÏ†ê Ï∂îÏ∂ú
    cart_times = category_df[category_df['event_type'] == 'cart'].groupby('user_id')['event_datetime'].min()
    purchase_times = category_df[category_df['event_type'] == 'purchase'].groupby('user_id')['event_datetime'].min()

    # cart ‚Üí purchase Ïó∞Í≤∞
    merged = pd.merge(cart_times, purchase_times, on='user_id', suffixes=('_cart', '_purchase'))

    # Î≥¥Ïú† ÏãúÍ∞Ñ Í≥ÑÏÇ∞ (Îã®ÏúÑ: ÏãúÍ∞Ñ)
    hold_hours = (merged['event_datetime_purchase'] - merged['event_datetime_cart']).dt.total_seconds() / 3600

    return hold_hours

# Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÌèâÍ∑† Cart Î≥¥Ïú† ÏãúÍ∞Ñ Ï†ÄÏû•
results = {}
for category in target_categories:
    one_time_hold = calculate_cart_hold_hours(one_time_target, category)
    multi_time_hold = calculate_cart_hold_hours(multi_time_target, category)

    results[category] = {
        '1Ìöå Íµ¨Îß§Ïûê ÌèâÍ∑†': round(one_time_hold.mean(), 2),
        'Îã§Ìöå Íµ¨Îß§Ïûê ÌèâÍ∑†': round(multi_time_hold.mean(), 2)
    }

# Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
cart_hold_summary_df = pd.DataFrame(results).T.reset_index().rename(columns={'index': 'Ïπ¥ÌÖåÍ≥†Î¶¨'})

# Í≤∞Í≥º Ï∂úÎ†•
print("Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ 1Ìöå vs Îã§Ìöå Íµ¨Îß§Ïûê ÌèâÍ∑† Cart Î≥¥Ïú† ÏãúÍ∞Ñ (Îã®ÏúÑ: ÏãúÍ∞Ñ)")
display(cart_hold_summary_df)
```

```python
# 1. Í∞ÄÍ≤© Íµ¨Í∞Ñ Ï†ïÏùò
bins = [0, 100, 200, 300, 400, 500, 1000, float('inf')]
labels = ['0~100', '100~200', '200~300', '300~400', '400~500', '500~1000', '1000+']
notop_one_df['price_bin'] = pd.cut(notop_one_df['price'], bins=bins, labels=labels, right=False)

# 2. Ïú†Ï†Ä Îã®ÏúÑ Ï§ëÎ≥µ Ï†úÍ±∞
view_df = notop_one_df[notop_one_df['event_type'] == 'view'][['user_id', 'product_id', 'price_bin']].drop_duplicates()
purchase_df = notop_one_df[notop_one_df['event_type'] == 'purchase'][['user_id', 'product_id', 'price_bin']].drop_duplicates()

# 3. Í∞ÄÍ≤© Íµ¨Í∞ÑÎ≥Ñ Ïú†Ï†Ä Ïàò ÏßëÍ≥Ñ (FutureWarning Î∞©ÏßÄ ÏúÑÌï¥ observed=True Ï∂îÍ∞Ä)
view_stats = view_df.groupby('price_bin', observed=True)['user_id'].nunique().reset_index()
view_stats.columns = ['price_bin', 'view_user_count']

purchase_stats = purchase_df.groupby('price_bin', observed=True)['user_id'].nunique().reset_index()
purchase_stats.columns = ['price_bin', 'purchase_user_count']

# 4. Î≥ëÌï© Î∞è Ï†ÑÌôòÏú® Í≥ÑÏÇ∞
price_group_stats = pd.merge(view_stats, purchase_stats, on='price_bin', how='outer')

# ‚û§ Ïà´Ïûê Ïª¨ÎüºÎßå fillna(0)
price_group_stats['view_user_count'] = price_group_stats['view_user_count'].fillna(0)
price_group_stats['purchase_user_count'] = price_group_stats['purchase_user_count'].fillna(0)

price_group_stats['conversion_rate'] = price_group_stats['purchase_user_count'] / price_group_stats['view_user_count']
price_group_stats = price_group_stats.sort_values(by='price_bin')

# 5. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
plt.bar(price_group_stats['price_bin'].astype(str), price_group_stats['conversion_rate'])
plt.title('Í∞ÄÍ≤© Íµ¨Í∞ÑÎ≥Ñ Ï†ÑÌôòÏú® (Ïú†Ï†Ä Í∏∞Ï§Ä)')
plt.xlabel('Í∞ÄÍ≤© Íµ¨Í∞Ñ (ÎßåÏõê)')
plt.ylabel('Ï†ÑÌôòÏú®')
plt.tight_layout()
plt.show()
```

### üîé Ïπ¥ÌÖåÍ≥†Î¶¨ Î†àÎ≤®3Î≥Ñ Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù(Ï†ÑÌôòÏú®)

```python
# 0. event_time ÏÉùÏÑ±
notop_one_df['event_time'] = pd.to_datetime(notop_one_df['event_date']) + pd.to_timedelta(notop_one_df['event_hour'], unit='h')

# 1. category_label ÏÉùÏÑ±
notop_one_df['category_label'] = notop_one_df['category_lvl1'] + ' > ' + notop_one_df['category_lvl3']

# 2. Ïú†Ï†Ä + Ïπ¥ÌÖåÍ≥†Î¶¨ Îã®ÏúÑÎ°ú Ïù¥Î≤§Ìä∏ Ï†ïÎ†¨
notop_one_df['event_datetime'] = notop_one_df['event_time']
notop_one_df_sorted = notop_one_df.sort_values(by=['user_id', 'category_label', 'event_datetime'])

# 3. Ïú†Ï†Ä Îã®ÏúÑÎ°ú ÏàúÏÑú Í∏∞Î∞ò Ï†ÑÌôòÏú® Í≥ÑÏÇ∞ Ìï®Ïàò Ï†ïÏùò
def calculate_ordered_conversion_rate(df):
    user_events = df.groupby(['user_id', 'category_label'])['event_type'].apply(list).reset_index()

    results = []

    for _, row in user_events.iterrows():
        events = row['event_type']
        category = row['category_label']

        view_to_cart = 0
        cart_to_purchase = 0
        view_to_purchase = 0

        if 'view' in events and 'cart' in events:
            if events.index('view') < events.index('cart'):
                view_to_cart = 1
        if 'cart' in events and 'purchase' in events:
            if events.index('cart') < events.index('purchase'):
                cart_to_purchase = 1
        if 'view' in events and 'purchase' in events:
            if events.index('view') < events.index('purchase'):
                view_to_purchase = 1

        results.append({
            'category_label': category,
            'view_to_cart': view_to_cart,
            'cart_to_purchase': cart_to_purchase,
            'view_to_purchase': view_to_purchase
        })

    result_df = pd.DataFrame(results)

    # Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ï†ÑÌôòÏú® ÌèâÍ∑† (Ïú†Ï†Ä Îã®ÏúÑ)
    conversion_stats = result_df.groupby('category_label').mean().reset_index()
    return conversion_stats

# 4. Ï†ÑÌôòÏú® Í≥ÑÏÇ∞
conversion_stats = calculate_ordered_conversion_rate(notop_one_df_sorted)

# 5. Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ïú†Ï†Ä Ïàò ÌïÑÌÑ∞ÎßÅ (30Î™Ö Ïù¥ÏÉÅ Î≥∏ Ïπ¥ÌÖåÍ≥†Î¶¨Îßå)
user_category_counts = notop_one_df.groupby(['category_label', 'user_id']).size().reset_index(name='event_count')
user_counts_per_category = user_category_counts.groupby('category_label')['user_id'].nunique().reset_index(name='user_count')

# Ï†ÑÌôòÏú® + Ïú†Ï†Ä Ïàò Í≤∞Ìï©
conversion_stats = conversion_stats.merge(user_counts_per_category, on='category_label', how='left')

# ÌïÑÌÑ∞ÎßÅ
conversion_stats_filtered = conversion_stats[conversion_stats['user_count'] >= 30]

# Ï†ïÎ†¨
top_funnel_sorted = conversion_stats_filtered.sort_values(by='cart_to_purchase', ascending=False).head(10)
top_funnel_sorted

# 6. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(14, 6))
bar_width = 0.25
x = range(len(top_funnel_sorted))

plt.bar(x, top_funnel_sorted['view_to_cart'], width=bar_width, label='View ‚Üí Cart')
plt.bar([p + bar_width for p in x], top_funnel_sorted['cart_to_purchase'], width=bar_width, label='Cart ‚Üí Purchase')
plt.bar([p + bar_width*2 for p in x], top_funnel_sorted['view_to_purchase'], width=bar_width, label='View ‚Üí Purchase')

plt.xticks([p + bar_width for p in x], top_funnel_sorted['category_label'], rotation=45)
plt.ylabel('Ï†ÑÌôòÏú® (Ïú†Ï†Ä Í∏∞Ï§Ä)')
plt.title('Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù (Ïú†Ï†Ä + Ïù¥Î≤§Ìä∏ ÏàúÏÑú Í∏∞Î∞ò)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 0. event_time ÏÉùÏÑ±
notop_multi_df['event_time'] = pd.to_datetime(notop_multi_df['event_date']) + pd.to_timedelta(notop_multi_df['event_hour'], unit='h')

# 1. category_label ÏÉùÏÑ±
notop_multi_df['category_label'] = notop_multi_df['category_lvl1'] + ' > ' + notop_multi_df['category_lvl3']

# 2. Ïú†Ï†Ä + Ïπ¥ÌÖåÍ≥†Î¶¨ Îã®ÏúÑÎ°ú Ïù¥Î≤§Ìä∏ Ï†ïÎ†¨
notop_multi_df['event_datetime'] = notop_multi_df['event_time']
notop_multi_df_sorted = notop_multi_df.sort_values(by=['user_id', 'category_label', 'event_datetime'])

# 3. Ïú†Ï†Ä Îã®ÏúÑÎ°ú ÏàúÏÑú Í∏∞Î∞ò Ï†ÑÌôòÏú® Í≥ÑÏÇ∞ Ìï®Ïàò Ï†ïÏùò
def calculate_ordered_conversion_rate(df):
    user_events = df.groupby(['user_id', 'category_label'])['event_type'].apply(list).reset_index()

    results = []

    for _, row in user_events.iterrows():
        events = row['event_type']
        category = row['category_label']

        view_to_cart = 0
        cart_to_purchase = 0
        view_to_purchase = 0

        if 'view' in events and 'cart' in events:
            if events.index('view') < events.index('cart'):
                view_to_cart = 1
        if 'cart' in events and 'purchase' in events:
            if events.index('cart') < events.index('purchase'):
                cart_to_purchase = 1
        if 'view' in events and 'purchase' in events:
            if events.index('view') < events.index('purchase'):
                view_to_purchase = 1

        results.append({
            'category_label': category,
            'view_to_cart': view_to_cart,
            'cart_to_purchase': cart_to_purchase,
            'view_to_purchase': view_to_purchase
        })

    result_df = pd.DataFrame(results)

    # Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ï†ÑÌôòÏú® ÌèâÍ∑† (Ïú†Ï†Ä Îã®ÏúÑ)
    conversion_stats = result_df.groupby('category_label').mean().reset_index()
    return conversion_stats

# 4. Ï†ÑÌôòÏú® Í≥ÑÏÇ∞
conversion_stats = calculate_ordered_conversion_rate(notop_multi_df_sorted)

# 5. Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ïú†Ï†Ä Ïàò ÌïÑÌÑ∞ÎßÅ (30Î™Ö Ïù¥ÏÉÅ Î≥∏ Ïπ¥ÌÖåÍ≥†Î¶¨Îßå)
user_category_counts = notop_multi_df.groupby(['category_label', 'user_id']).size().reset_index(name='event_count')
user_counts_per_category = user_category_counts.groupby('category_label')['user_id'].nunique().reset_index(name='user_count')

# Ï†ÑÌôòÏú® + Ïú†Ï†Ä Ïàò Í≤∞Ìï©
conversion_stats = conversion_stats.merge(user_counts_per_category, on='category_label', how='left')

# ÌïÑÌÑ∞ÎßÅ
conversion_stats_filtered = conversion_stats[conversion_stats['user_count'] >= 30]

# Ï†ïÎ†¨
top_funnel_sorted = conversion_stats_filtered.sort_values(by='cart_to_purchase', ascending=False).head(10)
top_funnel_sorted

# 6. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(14, 6))
bar_width = 0.25
x = range(len(top_funnel_sorted))

plt.bar(x, top_funnel_sorted['view_to_cart'], width=bar_width, label='View ‚Üí Cart')
plt.bar([p + bar_width for p in x], top_funnel_sorted['cart_to_purchase'], width=bar_width, label='Cart ‚Üí Purchase')
plt.bar([p + bar_width*2 for p in x], top_funnel_sorted['view_to_purchase'], width=bar_width, label='View ‚Üí Purchase')

plt.xticks([p + bar_width for p in x], top_funnel_sorted['category_label'], rotation=45)
plt.ylabel('Ï†ÑÌôòÏú® (Ïú†Ï†Ä Í∏∞Ï§Ä)')
plt.title('Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù (Ïú†Ï†Ä + Ïù¥Î≤§Ìä∏ ÏàúÏÑú Í∏∞Î∞ò)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 0. event_time ÏÉùÏÑ±
notop_multi_df['event_time'] = pd.to_datetime(notop_multi_df['event_date']) + pd.to_timedelta(notop_multi_df['event_hour'], unit='h')

# 1. categoryÎ≥Ñ ÏßëÍ≥Ñ (view ‚Üí cart / view ‚Üí purchase)
views = notop_multi_df[notop_multi_df['event_type'] == 'view']
carts = notop_multi_df[notop_multi_df['event_type'] == 'cart']
purchases = notop_multi_df[notop_multi_df['event_type'] == 'purchase']

# category label ÏÉùÏÑ±
views['category_label'] = views['category_lvl1'] + ' > ' + views['category_lvl3']
carts['category_label'] = carts['category_lvl1'] + ' > ' + carts['category_lvl3']
purchases['category_label'] = purchases['category_lvl1'] + ' > ' + purchases['category_lvl3']

view_counts = views.groupby('category_label').size().reset_index(name='view_count')
cart_counts = carts.groupby('category_label').size().reset_index(name='cart_count')
purchase_counts = purchases.groupby('category_label').size().reset_index(name='purchase_count')

category_stats = view_counts.merge(cart_counts, on='category_label', how='outer')
category_stats = category_stats.merge(purchase_counts, on='category_label', how='outer')
category_stats.fillna(0, inplace=True)

category_stats['view_to_cart_rate'] = category_stats['cart_count'] / category_stats['view_count']
category_stats['view_to_purchase_rate'] = category_stats['purchase_count'] / category_stats['view_count']

# 2. cart ‚Üí purchase Ï†ÑÌôòÏú® (user_id + category Í∏∞Ï§Ä)
carts_time = carts.sort_values('event_time').groupby(['user_id', 'category_label'], as_index=False).first()
purchases_time = purchases.sort_values('event_time').groupby(['user_id', 'category_label'], as_index=False).first()

cart_purchase = pd.merge(carts_time, purchases_time, on=['user_id', 'category_label'], suffixes=('_cart', '_purchase'))
cart_purchase = cart_purchase[cart_purchase['event_time_purchase'] > cart_purchase['event_time_cart']]
converted_cart_purchase = cart_purchase[['user_id', 'category_label']].drop_duplicates()
converted_cart_purchase['to_purchase'] = True

carts_time['to_purchase'] = carts_time.merge(
    converted_cart_purchase,
    on=['user_id', 'category_label'],
    how='left'
)['to_purchase'].fillna(False).astype(bool)

cart_to_purchase_rate = carts_time.groupby('category_label')['to_purchase'].mean().reset_index()
cart_to_purchase_rate.columns = ['category_label', 'cart_to_purchase_rate']

# 3. ÌÜµÌï©
category_stats = pd.merge(category_stats, cart_to_purchase_rate, on='category_label', how='left')
category_stats.fillna(0, inplace=True)

# 4. ÌïÑÌÑ∞ÎßÅ Î∞è Ï†ïÎ†¨
category_stats_filtered = category_stats[category_stats['view_count'] >= 30]
top_funnel_sorted = category_stats_filtered.sort_values(by='view_to_purchase_rate', ascending=False).head(10)

# 5. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(14, 6))
bar_width = 0.25
x = range(len(top_funnel_sorted))

plt.bar(x, top_funnel_sorted['view_to_cart_rate'], width=bar_width, label='View ‚Üí Cart')
plt.bar([p + bar_width for p in x], top_funnel_sorted['cart_to_purchase_rate'], width=bar_width, label='Cart ‚Üí Purchase')
plt.bar([p + bar_width*2 for p in x], top_funnel_sorted['view_to_purchase_rate'], width=bar_width, label='View ‚Üí Purchase')

plt.xticks([p + bar_width for p in x], top_funnel_sorted['category_label'], rotation=45)
plt.ylabel('Ï†ÑÌôòÏú®')
plt.title('Ï†ÑÌôò ÌçºÎÑê Î∂ÑÏÑù (Ïπ¥ÌÖåÍ≥†Î¶¨ Í∏∞Ï§Ä ÏßëÍ≥Ñ)')
plt.legend()
plt.tight_layout()
plt.show()
```

### üîé Ï≤´ Íµ¨Îß§ ÌñâÎèô Ìå®ÌÑ¥ Î∂ÑÏÑù(Ïû•Î∞îÍµ¨Îãà Î≥¥Ïú† ÏãúÍ∞Ñ, Ï¶âÏãú Íµ¨Îß§Ïûê ÎπÑÏú® ÌôïÏù∏)

```python
# Ï≤´ Íµ¨Îß§ Ïù¥Ï†Ñ cart -> purchase ÏãúÍ∞Ñ Í≥ÑÏÇ∞
# 1. event_datetime Ïª¨Îüº ÏÉùÏÑ±
notop_multi_df['event_datetime'] = notop_multi_df['event_date'] + pd.to_timedelta(notop_multi_df['event_hour'], unit='h')
notop_multi_df['event_datetime'] = pd.to_datetime(notop_multi_df['event_datetime'])

# 2. Í∞Å Ïú†Ï†ÄÏùò Ï≤´ purchase ÏãúÍ∞Ñ Íµ¨ÌïòÍ∏∞
first_purchase_time = notop_multi_df[notop_multi_df['event_type'] == 'purchase'] \
                        .groupby('user_id')['event_datetime'].min() \
                        .rename('first_purchase_time')

# 3. Ï≤´ Íµ¨Îß§ Ïù¥Ï†ÑÏùò ÌñâÎèôÎßå ÌïÑÌÑ∞ÎßÅ
notop_multi_df = notop_multi_df.join(first_purchase_time, on='user_id')
notop_multi_df_first_purchase = notop_multi_df[notop_multi_df['event_datetime'] <= notop_multi_df['first_purchase_time']].copy()

# 4. Ìï¥Îãπ Íµ¨Í∞Ñ ÎÇ¥ÏóêÏÑú cart/purchase ÏãúÏ†ê Ï∂îÏ∂ú
df_cart = notop_multi_df_first_purchase[notop_multi_df_first_purchase['event_type'] == 'cart']
df_purchase = notop_multi_df_first_purchase[notop_multi_df_first_purchase['event_type'] == 'purchase']

cart_times = df_cart.groupby('user_id')['event_datetime'].min()
purchase_times = df_purchase.groupby('user_id')['event_datetime'].min()

# 5. Í≥µÌÜµ user_id Í∏∞Ï§Ä Î≥ëÌï© Î∞è ÏãúÍ∞Ñ Ï∞® Í≥ÑÏÇ∞ (Îã®ÏúÑ: ÏãúÍ∞Ñ)
merged = pd.merge(cart_times, purchase_times, on='user_id', suffixes=('_cart', '_purchase'))
cart_hold_time_first_purchase = (merged['event_datetime_purchase'] - merged['event_datetime_cart']).dt.total_seconds() / 3600
```

```python
# Step 1: Íµ¨Îß§ ÌöüÏàò Í∏∞Î∞ò Í≥†Í∞ù Î∂ÑÎ¶¨
purchase_counts = df_other[df_other['event_type'] == 'purchase'].groupby('user_id').size()
one_time_buyers = purchase_counts[purchase_counts == 1].index
multi_buyers = purchase_counts[purchase_counts >= 2].index

# Step 2: 1Ìöå Í≥†Í∞ù Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú®
category_rate_one = (
    df_other[df_other['user_id'].isin(one_time_buyers)]
    .groupby('category_lvl1')['event_type']
    .apply(lambda x: (x == 'purchase').mean())
    .rename('1Ìöå Íµ¨Îß§')
)

# Step 3: 2Ìöå Ïù¥ÏÉÅ Í≥†Í∞ù Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú®
category_rate_multi = (
    df_other[df_other['user_id'].isin(multi_buyers)]
    .groupby('category_lvl1')['event_type']
    .apply(lambda x: (x == 'purchase').mean())
    .rename('2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§')
)

# Step 4: Î≥ëÌï©
category_rate_compare = pd.concat([category_rate_one, category_rate_multi], axis=1).fillna(0)

# Step 5: ÏãúÍ∞ÅÌôî
category_rate_compare.sort_values('1Ìöå Íµ¨Îß§', ascending=True).plot(
    kind='barh', figsize=(9, 6), color=['#a6c4e8', '#4068e0']
)
plt.title('Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú® (1Ìöå vs 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ Í≥†Í∞ù)')
plt.xlabel('Íµ¨Îß§Ïú®')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### üîé Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ viewÏàò ÎπÑÍµê(1Ìöå vs 2ÌöåÏù¥ÏÉÅ Íµ¨Îß§ Í≥†Í∞ù, ÎßâÎåÄ)

```python
# 1. Íµ¨Îß§ ÌöüÏàò Í≥ÑÏÇ∞
purchase_counts = df_other[df_other['event_type'] == 'purchase'].groupby('user_id').size()

# 2. Í∑∏Î£π Î∂ÑÎ•ò
one_time_buyers = purchase_counts[purchase_counts == 1].index
multi_buyers    = purchase_counts[purchase_counts >= 2].index

# 1Ìöå Íµ¨Îß§ Í≥†Í∞ù view Ïàò
view_counts_one = (
    df_other[(df_other['event_type'] == 'view') & (df_other['user_id'].isin(one_time_buyers))]
    .groupby('category_lvl1')['user_id']
    .count()
)

# 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ Í≥†Í∞ù view Ïàò
view_counts_multi = (
    df_other[(df_other['event_type'] == 'view') & (df_other['user_id'].isin(multi_buyers))]
    .groupby('category_lvl1')['user_id']
    .count()
)

# Í≤∞Ìï© (Îπà Ïπ¥ÌÖåÍ≥†Î¶¨ Ï±ÑÏö∞Í∏∞ ÏúÑÌï¥ outer join)
view_compare = pd.concat(
    [view_counts_one.rename('1Ìöå Íµ¨Îß§'), view_counts_multi.rename('2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§')],
    axis=1
).fillna(0).astype(int).sort_values('1Ìöå Íµ¨Îß§', ascending=False)

# ÎßâÎåÄÍ∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
view_compare.plot(kind='barh', figsize=(10, 6), color=['#a6c4e8', '#4068e0'])
plt.title('Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ View Ïàò ÎπÑÍµê (1Ìöå vs 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ Í≥†Í∞ù)')
plt.xlabel('View Ïàò')
plt.ylabel('Ïπ¥ÌÖåÍ≥†Î¶¨')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### üîé 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ÏûêÏùò Ï≤´->Îëê Î≤àÏß∏ Íµ¨Îß§ÏóêÏÑú Ïπ¥ÌÖåÍ≥†Î¶¨ Î≥ÄÍ≤Ω ÎπÑÏú®, Í≤ΩÎ°ú

```python
# Íµ¨Îß§ Ïù¥Î≤§Ìä∏Îßå ÌïÑÌÑ∞ÎßÅ
multi_purchase = notop_multi_df[notop_multi_df['event_type'] == 'purchase'].copy()

# ÎÇ†Ïßú Ï†ïÎ†¨
multi_purchase['event_date'] = pd.to_datetime(multi_purchase['event_date'])
multi_purchase = multi_purchase.sort_values(by=['user_id', 'event_date'])

# Ïú†Ï†ÄÎ≥Ñ Íµ¨Îß§ ÏàúÏÑú Î∂ÄÏó¨
multi_purchase['purchase_order'] = multi_purchase.groupby('user_id').cumcount() + 1

# Ï≤´ Î≤àÏß∏, Îëê Î≤àÏß∏ Íµ¨Îß§Îßå Ï∂îÏ∂ú
first_second = multi_purchase[multi_purchase['purchase_order'] <= 2]

# Ï≤´/Îëê Î≤àÏß∏ Íµ¨Îß§Î•º ÌïòÎÇòÏùò ÌñâÏóê ÌîºÎ≤ó
pivoted = first_second.pivot(index='user_id', columns='purchase_order', values='category_lvl1')
pivoted.columns = ['first_cat1', 'second_cat1']
pivoted = pivoted.dropna()

# Î≥ÄÍ≤Ω Ïó¨Î∂Ä ÌåêÎã®
pivoted['category_changed'] = pivoted['first_cat1'] != pivoted['second_cat1']

# ÌÜµÍ≥Ñ ÏöîÏïΩ
change_rate = pivoted['category_changed'].mean()
change_counts = pivoted['category_changed'].value_counts()

print("2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ÏûêÏùò Ï≤´‚ÜíÎëêÎ≤àÏß∏ Íµ¨Îß§ÏóêÏÑú Ïπ¥ÌÖåÍ≥†Î¶¨ Î≥ÄÍ≤Ω ÎπÑÏú®:", round(change_rate, 3))
print("\nÏÉÅÏÑ∏ Î∂ÑÌè¨:")
print(change_counts)

# Î≥ÄÍ≤ΩÎêú Í≤ΩÏö∞Îßå Ï∂îÏ∂ú
changed_cases = pivoted[pivoted['category_changed']]

# Î≥ÄÌôî Í≤ΩÎ°ú ÏßëÍ≥Ñ
change_paths = changed_cases.groupby(['first_cat1', 'second_cat1']).size().reset_index(name='count')
change_paths = change_paths.sort_values('count', ascending=False)

# ÏÉÅÏúÑ Î™á Í∞úÎßå ÏãúÍ∞ÅÌôî
top_paths = change_paths.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_paths, y='count', x='first_cat1', hue='second_cat1')
plt.title('2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ÏûêÏùò Ïπ¥ÌÖåÍ≥†Î¶¨ Î≥ÄÍ≤Ω Í≤ΩÎ°ú (Top 10)')
plt.xlabel('Ï≤´ Î≤àÏß∏ Íµ¨Îß§ Ïπ¥ÌÖåÍ≥†Î¶¨')
plt.ylabel('Î≥ÄÍ≤ΩÎêú Ïàò')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### üîé 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ÏûêÏùò Ïπ¥ÌÖåÍ≥†Î¶¨ Ï°∞Ìï©

```python
# 1. Íµ¨Îß§ Ïù¥Î≤§Ìä∏ ÌïÑÌÑ∞ÎßÅ
multi_purchase = notop_multi_df[notop_multi_df['event_type'] == 'purchase']

# 2. category_lvl2 ‚Üí category_lvl1 Îß§Ìïë ÌÖåÏù¥Î∏î ÏÉùÏÑ±
lvl2_lvl1_map = multi_purchase[['category_lvl2', 'category_lvl1']].drop_duplicates()

# 3. Ïú†Ï†ÄÎ≥Ñ Íµ¨Îß§Ìïú category_lvl2 Ï°∞Ìï© ÏÉùÏÑ±
cat_combo = multi_purchase.groupby('user_id')['category_lvl2'].apply(lambda x: tuple(sorted(set(x))))
cat_combo = cat_combo.value_counts().head(10).reset_index()
cat_combo.columns = ['category_combination', 'user_count']

# 4. Í∞Å Ï°∞Ìï©Ïóê Ìè¨Ìï®Îêú category_lvl1 Ï†ïÎ≥¥ Ï∂îÍ∞Ä
def map_to_lvl1(combo):
    lvl1_set = lvl2_lvl1_map[lvl2_lvl1_map['category_lvl2'].isin(combo)]['category_lvl1'].unique()
    return ', '.join(sorted(set(lvl1_set)))

cat_combo['category_lvl1'] = cat_combo['category_combination'].apply(map_to_lvl1)

# print(cat_combo)

## ÏãúÍ∞ÅÌôî ##
# 5. ÌäúÌîåÏùÑ Î¨∏ÏûêÏó¥Î°ú Î≥ÄÌôò (ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌï¥)
cat_combo['category_combination'] = cat_combo['category_combination'].apply(lambda x: ', '.join(x))

# 6. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12, 6))
sns.barplot(data=cat_combo, y='category_combination', x='user_count')
plt.title('2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ÏûêÏùò Ïπ¥ÌÖåÍ≥†Î¶¨ Ï°∞Ìï© Top 10 (LV2 Í∏∞Ï§Ä)')
plt.xlabel('Ïú†Ï†Ä Ïàò')
plt.ylabel('LV2 Ï°∞Ìï© (‚Üí LV1: ' + cat_combo['category_lvl1'].iloc[0] + ' Îì±)')
plt.tight_layout()
plt.show()

# Í≤∞Í≥º ÌôïÏù∏
display(cat_combo[['category_combination', 'user_count', 'category_lvl1']])
```

### üîé Ïû¨Íµ¨Îß§Ïú® ÎÇÆÏùÄ Ïπ¥ÌÖåÍ≥†Î¶¨ 2

```python
# notop_one_df: Ìïú Î≤àÎßå Íµ¨Îß§Ìïú Ïú†Ï†ÄÎì§
# notop_multi_df: Îëê Î≤à Ïù¥ÏÉÅ Íµ¨Îß§Ìïú Ïú†Ï†ÄÎì§
# => Îëê ÏßëÌï©ÏùÄ Ïú†Ï†Ä Í∏∞Ï§ÄÏúºÎ°ú ÏôÑÏ†ÑÌûà Î∂ÑÎ¶¨Îê® ‚Üí Ï§ëÎ≥µ ÏóÜÏùå
# Îî∞ÎùºÏÑú (repeat_users / (total_users + repeat_users))Îäî Ï†ïÌôïÌïú Ïû¨Íµ¨Îß§Ïú® Í≥ÑÏÇ∞ÏûÑ

# Ï†ÑÏ≤¥ Ïú†Ï†Ä ÎåÄÎπÑ Ïû¨Íµ¨Îß§ Ïú†Ï†Ä ÎπÑÏú®
total_users = notop_one_df[notop_one_df['event_type'] == 'purchase'].groupby('category_lvl2')['user_id'].nunique()
repeat_users = notop_multi_df[notop_multi_df['event_type'] == 'purchase'].groupby('category_lvl2')['user_id'].nunique()

repurchase_rate = (repeat_users / (total_users + repeat_users)).reset_index()
repurchase_rate.columns = ['category_lvl2', 'repurchase_rate']

# ÌïòÏúÑ 10Í∞ú Ïπ¥ÌÖåÍ≥†Î¶¨ ÏãúÍ∞ÅÌôî
bottom_10 = repurchase_rate.sort_values(by='repurchase_rate').head(10)

plt.figure(figsize=(8,5))
sns.barplot(data=bottom_10, x='repurchase_rate', y='category_lvl2')
plt.title('Ïû¨Íµ¨Îß§Ïú®Ïù¥ ÎÇÆÏùÄ Ïπ¥ÌÖåÍ≥†Î¶¨ Top10')
plt.xlabel('Ïû¨Íµ¨Îß§Ïú®')
plt.ylabel('Ïπ¥ÌÖåÍ≥†Î¶¨')
plt.tight_layout()
plt.show()
```

```python
df = notop_one_df  # 1Ìöå Íµ¨Îß§ÏûêÏö© Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞

# 1. event_datetime ÏÉùÏÑ± (vectorized Î∞©Ïãù)
df['event_datetime'] = pd.to_datetime(df['event_date']) + pd.to_timedelta(df['event_hour'], unit='h')

# 2. cart & purchase Ïù¥Î≤§Ìä∏ ÌïÑÌÑ∞ÎßÅ
df_cart = df[df['event_type'] == 'cart']
df_purchase = df[df['event_type'] == 'purchase']

# 3. ÏÑ∏ÏÖòÎ≥Ñ Í∞ÄÏû• Îπ†Î•∏ cart/purchase ÏãúÍ∞Ñ Ï∂îÏ∂ú
cart_times = df_cart.groupby('user_session')['event_datetime'].min().to_frame(name='cart_time')
purchase_times = df_purchase.groupby('user_session')['event_datetime'].min().to_frame(name='purchase_time')

# 4. Í≥µÌÜµ ÏÑ∏ÏÖòÎßå ÎÇ®Í∏∞Í≥† merge
merged = pd.merge(cart_times, purchase_times, left_index=True, right_index=True)

# 5. ÏãúÍ∞ÑÏ∞® Í≥ÑÏÇ∞
cart_hold_time = (merged['purchase_time'] - merged['cart_time']).dt.total_seconds() / 3600

# 6. ÏùåÏàò Ï†úÍ±∞ (Ïò¨Î∞îÎ•∏ ÏàúÏÑúÎßå ÌïÑÌÑ∞ÎßÅ)
cart_hold_time = cart_hold_time[cart_hold_time >= 0]

# 7. Í≤∞Í≥º Ï∂úÎ†•
print(f"[1Ìöå Íµ¨Îß§Ïûê] ÌèâÍ∑† Cart Î≥¥Ïú† ÏãúÍ∞Ñ (ÏãúÍ∞Ñ): {cart_hold_time.mean():.2f}")
```
