## üö©ÎπÑÏ£ºÎ†• ÏÉÅÌíà Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú® ÌôïÏù∏ÌïòÍ∏∞

```python
# ‚ñ£ Step 1: Íµ¨Îß§ ÌöüÏàò Í∏∞Î∞ò Í≥†Í∞ù Î∂ÑÎ¶¨
purchase_counts = df[df['event_type'] == 'purchase'].groupby('user_id').size()
one_time_buyers = purchase_counts[purchase_counts == 1].index
multi_buyers = purchase_counts[purchase_counts >= 2].index

# ‚ñ£ Step 2: 1Ìöå Í≥†Í∞ù Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú®
category_rate_one = (
    df[df['user_id'].isin(one_time_buyers)]
    .groupby('category_lvl1')['event_type']
    .apply(lambda x: (x == 'purchase').mean())
    .rename('1Ìöå Íµ¨Îß§')
)

# ‚ñ£ Step 3: 2Ìöå Ïù¥ÏÉÅ Í≥†Í∞ù Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú®
category_rate_multi = (
    df[df['user_id'].isin(multi_buyers)]
    .groupby('category_lvl1')['event_type']
    .apply(lambda x: (x == 'purchase').mean())
    .rename('2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§')
)

# ‚ñ£ Step 4: Î≥ëÌï©
category_rate_compare = pd.concat([category_rate_one, category_rate_multi], axis=1).fillna(0)

# ‚ñ£ Step 5: ÏãúÍ∞ÅÌôî
category_rate_compare.sort_values('1Ìöå Íµ¨Îß§', ascending=True).plot(
    kind='barh', figsize=(9, 6), color=['#a6c4e8', '#4068e0']
)
plt.title('Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Íµ¨Îß§Ïú® (1Ìöå vs 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ Í≥†Í∞ù)')
plt.xlabel('Íµ¨Îß§Ïú®')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## üö©ÎπÑÏ£ºÎ†• ÏÉÅÌíàÎ≥Ñ view Ïàò ÌôïÏù∏ÌïòÍ∏∞

```python
# 1. Íµ¨Îß§ ÌöüÏàò Í≥ÑÏÇ∞
purchase_counts = df[df['event_type'] == 'purchase'].groupby('user_id').size()

# 2. Í∑∏Î£π Î∂ÑÎ•ò
one_time_buyers = purchase_counts[purchase_counts == 1].index
multi_buyers    = purchase_counts[purchase_counts >= 2].index

# 1Ìöå Íµ¨Îß§ Í≥†Í∞ù view Ïàò
view_counts_one = (
    df[(df['event_type'] == 'view') & (df['user_id'].isin(one_time_buyers))]
    .groupby('category_lvl1')['user_id']
    .count()
)

# 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ Í≥†Í∞ù view Ïàò
view_counts_multi = (
    df[(df['event_type'] == 'view') & (df['user_id'].isin(multi_buyers))]
    .groupby('category_lvl1')['user_id']
    .count()
)

# Í≤∞Ìï© (Îπà Ïπ¥ÌÖåÍ≥†Î¶¨ Ï±ÑÏö∞Í∏∞ ÏúÑÌï¥ outer join)
view_compare = pd.concat(
    [view_counts_one.rename('1Ìöå Íµ¨Îß§'), view_counts_multi.rename('2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§')],
    axis=1
).fillna(0).astype(int).sort_values('1Ìöå Íµ¨Îß§', ascending=False)

# ÎßâÎåÄÍ∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
view_compare.plot(kind='barh', figsize=(10, 6), color=['#a6c4e8', '#4068e0'])
plt.title('Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ View Ïàò ÎπÑÍµê (1Ìöå vs 2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§ Í≥†Í∞ù)')
plt.xlabel('View Ïàò')
plt.ylabel('Ïπ¥ÌÖåÍ≥†Î¶¨')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Íµ¨Îß§ ÌöüÏàò Í∏∞Ï§ÄÏúºÎ°ú Í≥†Í∞ù Î∂ÑÎ¶¨
purchase_counts = df[df['event_type'] == 'purchase'].groupby('user_id').size()
one_time_buyers = purchase_counts[purchase_counts == 1].index
multi_buyers = purchase_counts[purchase_counts >= 2].index

df_one = df[df['user_id'].isin(one_time_buyers)]
df_multi = df[df['user_id'].isin(multi_buyers)]

# Ìï®Ïàò: view ‚Üí purchaseÍπåÏßÄ ÏÜåÏöî ÏãúÍ∞Ñ (Î∂Ñ Îã®ÏúÑ)
def compute_time_to_purchase(df_group):
    view_time = df_group[df_group['event_type'] == 'view'].groupby('user_id')['event_time'].min()
    purchase_time = df_group[df_group['event_type'] == 'purchase'].groupby('user_id')['event_time'].min()
    df_time = pd.merge(view_time, purchase_time, on='user_id', how='inner', suffixes=('_view', '_purchase'))
    df_time['diff_min'] = (df_time['event_time_purchase'] - df_time['event_time_view']).dt.total_seconds() / 60
    return df_time['diff_min']

# Í∞Å Í∑∏Î£πÎ≥Ñ ÏÜåÏöî ÏãúÍ∞Ñ Í≥ÑÏÇ∞
time_one = compute_time_to_purchase(df_one)
time_multi = compute_time_to_purchase(df_multi)

# ‚ñ£ 1. ÏùåÏàò Ï†úÍ±∞ (viewÎ≥¥Îã§ purchaseÍ∞Ä Î®ºÏ†ÄÏù∏ Ïò§Î•ò Ï†úÍ±∞)
time_one = time_one[time_one > 0]
time_multi = time_multi[time_multi > 0]

# ‚ñ£ 2. Í∑πÎã®Ïπò Ï†úÍ±∞ (1Ïùº = 1440Î∂Ñ Ïù¥ÌïòÎßå ÎÇ®ÍπÄ)
time_one = time_one[time_one <= 1440]
time_multi = time_multi[time_multi <= 1440]

# ‚ñ£ 3. KDE ÏãúÍ∞ÅÌôî
plt.figure(figsize=(8, 5))
sns.kdeplot(time_one, label='1Ìöå Íµ¨Îß§', fill=True, color='#a6c4e8')
sns.kdeplot(time_multi, label='2Ìöå Ïù¥ÏÉÅ Íµ¨Îß§', fill=True, color='#e0a3a3')
plt.title('Íµ¨Îß§ÍπåÏßÄ ÏÜåÏöî ÏãúÍ∞Ñ Î∂ÑÌè¨ (1Ïùº Ïù¥ÎÇ¥, Î∂Ñ Îã®ÏúÑ)')
plt.xlabel('ÏÜåÏöî ÏãúÍ∞Ñ (Î∂Ñ)')
plt.ylabel('Î∞ÄÎèÑ')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from tqdm import tqdm
import numpy as np

# ‚ñ£ 1. Î™©Ìëú ÏÉòÌîå Ïàò ÏÑ§Ï†ï
target_total_rows = 7_500_000

# ‚ñ£ 2. Ï∏µÌôî Í∏∞Ï§Ä Ïó¥ ÏÉùÏÑ±
df['stratum'] = (
    df['category_lvl1'].fillna('Unknown') + ' > ' +
    df['category_lvl2'].fillna('Unknown') + ' | ' +
    df['event_type']
)

# ‚ñ£ 3. Î™®ÏßëÎã® ÎÇ¥ Í∞Å Í∑∏Î£π ÎπÑÏú® Í≥ÑÏÇ∞
group_counts = df['stratum'].value_counts()
group_ratio = group_counts / group_counts.sum()

# ‚ñ£ 4. Í∞Å Í∑∏Î£πÎ≥Ñ ÏÉòÌîå Ïàò Í≥ÑÏÇ∞ (ÎÇ¥Î¶º Ï≤òÎ¶¨)
raw_sizes = group_ratio * target_total_rows
group_sample_sizes = np.floor(raw_sizes).astype(int)

# ‚ñ£ 5. ÎÇ®ÏùÄ ÏàòÎßåÌÅº ÏÜåÏàòÏ†ê Í∏∞Ï§ÄÏúºÎ°ú Î∞∞Î∂Ñ
gap = target_total_rows - group_sample_sizes.sum()
remainder = (raw_sizes - group_sample_sizes).sort_values(ascending=False)
for stratum in remainder.index[:gap]:
    group_sample_sizes[stratum] += 1

# ‚ñ£ 6. Í∑∏Î£πÌïë ÎØ∏Î¶¨ Ï§ÄÎπÑ
df_grouped = df.groupby('stratum')

# ‚ñ£ 7. ÏÉòÌîåÎßÅ ÏàòÌñâ (tqdmÏúºÎ°ú ÏßÑÌñâÎ•† ÌëúÏãú)
sampled_list = []
for stratum, n in tqdm(group_sample_sizes.items(), desc='ÏÉòÌîåÎßÅ Ï§ë'):
    if stratum in df_grouped.groups:
        df_stratum = df_grouped.get_group(stratum)
        n_sample = min(n, len(df_stratum))  # Îç∞Ïù¥ÌÑ∞ Î∂ÄÏ°± ÎåÄÎπÑ
        df_sampled = df_stratum.sample(n=n_sample, random_state=42)
        sampled_list.append(df_sampled)

# ‚ñ£ 8. ÏµúÏ¢Ö Î≥ëÌï©
df_stratified_sample = pd.concat(sampled_list).reset_index(drop=True)

# ‚ñ£ 9. Í≤∞Í≥º ÌôïÏù∏
print(f"ÏµúÏ¢Ö ÏÉòÌîå Ïàò: {len(df_stratified_sample):,}Í∞ú")
```

```python
import pandas as pd
from tqdm import tqdm

# ‚ñ£ 0. Ï†ÑÏ≤¥ row Ïàò
total_rows = 50_000_000
target_rows = 7_500_000
sample_ratio = target_rows / total_rows

# ‚ñ£ 1. strata Ïª¨Îüº ÏÉùÏÑ±
df['strata'] = (
    df['category_lvl1'].astype(str) + ' | ' +
    df['event_type'].astype(str) + ' | ' +
    df['weekday'].astype(str) + ' | ' +
    df['event_hour'].astype(str)  # or ‚Üí df['event_hour'] // 6 ‚Üí ÏãúÍ∞ÑÎåÄÎ≥Ñ
)

# ‚ñ£ 2. strata ÎπÑÏú® Í≥ÑÏÇ∞
strata_counts = df['strata'].value_counts()
strata_ratios = strata_counts / strata_counts.sum()

# ‚ñ£ 3. strataÎ≥Ñ ÏÉòÌîå Ïàò Í≥ÑÏÇ∞
strata_sizes = (strata_ratios * target_rows).astype(int)

# ‚ñ£ 4. strataÎ≥Ñ ÏÉòÌîåÎßÅ
sampled_list = []

for strata_val, n_samples in tqdm(strata_sizes.items()):
    temp = df[df['strata'] == strata_val]
    if len(temp) >= n_samples:
        sampled = temp.sample(n=n_samples, random_state=42)
    else:
        sampled = temp  # ÎÑàÎ¨¥ Ï†ÅÏùÄ Í≤ΩÏö∞ Ï†ÑÏ≤¥ ÏÇ¨Ïö©
    sampled_list.append(sampled)

df_sampled = pd.concat(sampled_list).reset_index(drop=True)
```

```python
# ÏõêÎ≥∏ vs ÏÉòÌîåÏùò Î∂ÑÌè¨ ÎπÑÍµê
def plot_compare_dist(col):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    df[col].value_counts(normalize=True).sort_index().plot(kind='bar', ax=axes[0], title='Original')
    df_sampled[col].value_counts(normalize=True).sort_index().plot(kind='bar', ax=axes[1], title='Sampled')
    plt.suptitle(f"{col} Î∂ÑÌè¨ ÎπÑÍµê")
    plt.tight_layout()
    plt.show()

plot_compare_dist('event_type')
plot_compare_dist('category_lvl1')
plot_compare_dist('event_hour')
plot_compare_dist('weekday')
```

```python
import pandas as pd

# 1. Íµ¨Îß§ Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂ú
purchase_df = df[df['event_type'] == 'purchase'].copy()
purchase_df['event_time'] = pd.to_datetime(purchase_df['event_time'])

# 2. Íµ¨Îß§ ÏãúÍ∞Ñ Ï†ïÎ†¨
purchase_df = purchase_df.sort_values(['user_id', 'category_lvl1', 'event_time'])

# 3. Í≥†Í∞ù+Ïπ¥ÌÖåÍ≥†Î¶¨ Îã®ÏúÑ Ïû¨Íµ¨Îß§ Í∞ÑÍ≤© Í≥ÑÏÇ∞
purchase_df['prev_time'] = purchase_df.groupby(['user_id', 'category_lvl1'])['event_time'].shift(1)
purchase_df['gap_days'] = (purchase_df['event_time'] - purchase_df['prev_time']).dt.days

# 4. Ï≤´ Íµ¨Îß§Îäî Ï†úÏô∏
gap_df = purchase_df.dropna(subset=['gap_days']).copy()

# 5. Í≥†Í∞ùÎ≥Ñ ÌèâÍ∑† Ïû¨Íµ¨Îß§ Í∞ÑÍ≤© Í≥ÑÏÇ∞ (categoryÎ≥Ñ)
user_cat_gap = gap_df.groupby(['user_id', 'category_lvl1'])['gap_days'].mean().reset_index()
user_cat_gap.rename(columns={'gap_days': 'avg_repurchase_days'}, inplace=True)

# 6. Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Î∂ÑÏúÑ Í∏∞Ï§ÄÏúºÎ°ú Î∂ÑÎ•ò
def classify_by_quantile(sub_df):
    q1 = sub_df['avg_repurchase_days'].quantile(0.33)
    q2 = sub_df['avg_repurchase_days'].quantile(0.66)

    # Î∂ÑÏúÑÏàòÍ∞Ä Í∞ôÏúºÎ©¥ 'Î≥¥ÌÜµ'ÏúºÎ°ú ÌÜµÏùº
    if q1 == q2:
        return pd.Series(['Î≥¥ÌÜµ'] * len(sub_df), index=sub_df.index)

    return pd.cut(
        sub_df['avg_repurchase_days'],
        bins=[-float('inf'), q1, q2, float('inf')],
        labels=['ÏûêÏ£º', 'Î≥¥ÌÜµ', 'ÎìúÎ¨∏']
    )


user_cat_gap['repurchase_group'] = user_cat_gap.groupby('category_lvl1', group_keys=False).apply(classify_by_quantile)

# 7. ÌôïÏù∏
print(user_cat_gap.head())
```

```python
import matplotlib.pyplot as plt

plot_df = user_cat_gap.groupby(['category_lvl1', 'repurchase_group']).size().unstack().fillna(0)
plot_df = plot_df.divide(plot_df.sum(axis=1), axis=0)  # ÎπÑÏú®Ìôî

plot_df.plot(kind='barh', stacked=True, figsize=(10, 6), colormap='Blues')
plt.title('Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ïû¨Íµ¨Îß§ Ï£ºÍ∏∞ Í∑∏Î£π ÎπÑÏú®')
plt.xlabel('ÎπÑÏú®')
plt.legend(title='Ïû¨Íµ¨Îß§ Ï£ºÍ∏∞')
plt.tight_layout()
plt.show()
```

```python
# ‚úÖ 1. Íµ¨Îß§ Ïù¥Î≤§Ìä∏ ÌïÑÌÑ∞ÎßÅ
df_purchase = df[df['event_type'] == 'purchase'].copy()
df_purchase['event_time'] = pd.to_datetime(df_purchase['event_time'])

# ‚úÖ 2. Ï†ïÎ†¨
df_purchase = df_purchase.sort_values(by=['user_id', 'category_lvl1', 'event_time'])

# ‚úÖ 3. Îã§Ïùå Íµ¨Îß§Ïùº Í≥ÑÏÇ∞
df_purchase['next_purchase'] = df_purchase.groupby(['user_id', 'category_lvl1'])['event_time'].shift(-1)

# ‚úÖ 4. Íµ¨Îß§ Í∞ÑÍ≤© Í≥ÑÏÇ∞
df_purchase['gap_days'] = (df_purchase['next_purchase'] - df_purchase['event_time']).dt.days
df_valid_gap = df_purchase[df_purchase['gap_days'].notna()].copy()

# ‚úÖ 5. Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ ÎàÑÏ†ÅÎ∂ÑÌè¨ Í∏∞Î∞ò Îì±Í∏â ÎùºÎ≤®ÎßÅ Ìï®Ïàò Ï†ïÏùò
def assign_repurchase_percentile(df):
    result = df.copy()
    # gap_days ÎàÑÏ†Å ÎπÑÏú® Í≥ÑÏÇ∞
    result['percentile'] = result['gap_days'].rank(pct=True)

    # Í∏∞Ï§Ä Íµ¨Í∞Ñ
    def label(p):
        if p <= 0.33:
            return 'ÏûêÏ£º'
        elif p <= 0.67:
            return 'Î≥¥ÌÜµ'
        else:
            return 'ÎìúÎ¨∏'

    result['revisit_type'] = result['percentile'].apply(label)
    return result

# ‚úÖ 6. Ïπ¥ÌÖåÍ≥†Î¶¨Î≥ÑÎ°ú Ï†ÅÏö©
df_percentiled = (
    df_valid_gap
    .groupby('category_lvl1', group_keys=False)
    .apply(assign_repurchase_percentile)
)

# ‚úÖ 7. Í≤∞Í≥º Î∂ÑÌè¨ ÌôïÏù∏Ïö© ÌîºÎ≤ó
pivot = df_percentiled.groupby(['category_lvl1', 'revisit_type'])['user_id'].count().unstack(fill_value=0)
pivot_percent = pivot.div(pivot.sum(axis=1), axis=0) * 100

# ‚úÖ 8. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
pivot_percent[['ÏûêÏ£º', 'Î≥¥ÌÜµ', 'ÎìúÎ¨∏']].plot(
    kind='barh', stacked=True, color=['#4068e0', '#a6c4e8', '#d8e0ea'], figsize=(10, 6)
)
plt.title("Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Ïû¨Íµ¨Îß§ Ïú†Ìòï Î∂ÑÌè¨ (ÌçºÏÑºÌÉÄÏùº Í∏∞Î∞ò)")
plt.xlabel("ÎπÑÏú® (%)")
plt.legend(title="Ïû¨Íµ¨Îß§ Ïú†Ìòï")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
